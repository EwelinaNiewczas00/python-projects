```python
def adventure():
    print("You staand at a fork in the road")
    choice = input("Do ypu go left or right? ")

    if choice == "left":
        print("You encounter a friendly dragon!")
        second = input("Do you talk to the dragon or walk away?")
        if second == "talk":
            print("The dragon goves you a magical map!")
        elif second == "walk away":
            print("You safely head back home, a  little disappointed")
        else:
            print("The dragon flies off while you hesistate")
    elif choice == "right":
        print("You find a hidden treasure chest!")
    else:
        print("You get lost in the woods")

adventure()
```

Słowniki / Dictionaries

```python
person = {
    "name" : "Ewa",
    "age" : "25",
    "city" : "Warsaw"
}
person["job"] = "postofficer"

print(person["name"])
print(person["age"])
print(person["city"])

for key, value in person.items():
    print(key, ":", value)
    print(f"{key} : {value}")
```

```python
people = [
    {"name" : "Ava", "age":25},
    {"name": "Zana", "age" :30},
    {"name": "maya", "age" :34}
]

for person in people:
    name = person.get("name")
    age = person.get("age")
    print(f" {name} is {age} years old")
print()

for person in people:
    name, age = person.values()
    print(f" {name} is {age} years old")
print()

for person in people:
    print(f" {person["name"]} is {person["age"]} years old")
```

Write to a file / read, append text files

```python
with open("message.txt", "w") as file:
    file.write("Hello!")

with open("message.txt", "r") as file:
    content = file.read()
    print(content)

with open("message.txt", "a") as file:
    file.write("\nCześć")

with open("message.txt", "r") as file:
    read = file.read()
    print(read)
```

```python
entry = input("Write a diary entry: ")

with open("diary.txt", "a") as file:
    file.write(entry + "\n")

with open("diary.txt", "r") as file:
    read = file.read()
    print(read)
```
